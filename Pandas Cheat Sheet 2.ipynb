```python
in1 = ipl[ipl['inning']==1]
in2 = ipl[ipl['inning']==2]
```

```python
in1 = in1.groupby('match')['total_runs'].sum().reset_index()
```

```python
in2 = in2.groupby('match')['total_runs'].sum().reset_index()
```

```python
in1 = in1[in1['total_runs']>=200]
```

```python
inning_score_data = pd.merge(in1,in2,on="match",how="inner")
```

```python
inning_score_data['result'] = inning_score_data['total_runs_x'] < inning_score_data['total_runs_y']
```

```python
inning_score_data['result'].value_counts()
```
