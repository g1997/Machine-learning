```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [11,12,15,18,21]

plt.plot(x,y)
plt.show()
```

```python
# Plot size
plt.figure(figsize=(10,10))
plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.plot(x,y)
plt.title("Price vs frequency distribution")
plt.xlabel("price")
plt.ylabel("frequency")
plt.show()
```

```python
# range
plt.figure(figsize=(7,7))
plt.plot(x,y)
plt.xlim((1,8))
plt.ylim((1,30))
plt.show()
```

```python
x = [-3,-4,-1,0,5,6,7]
y = [-9,-6,0,2,3,4,6]
plt.plot(x,y)
plt.show()
```

```python
x = [1,5,10,15,20]
y1 = [11,14,16,17,19]
y2 = [12,13,18,20,21]

plt.figure(figsize=(15,10))
plt.plot(x,y1,color="yellow")
plt.plot(x,y2,color="blue")
plt.legend(labels=['Y1','Y2'])
plt.show()
```

```python
x = [1,2,3,4,5]
y = [11,12,15,18,21]


plt.figure(figsize=(7,7))
plt.scatter(x,y)
plt.title("Price vs frequency distribution")
plt.xlabel("price")
plt.ylabel("frequency")
plt.show()
```

```python
x = ["A","B","C","D","E"]
y = [11,12,15,18,21]

plt.figure(figsize=(7,7))
plt.pie(y,labels=x)
plt.title("Price vs frequency distribution")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [11,12,15,18,21]

plt.subplot(2,1,1)
plt.plot(x,y)
plt.subplot(2,1,2)
plt.scatter(x,y)
plt.show()
```

```python
house = pd.read_csv("D://house.csv")
```

```python
plt.figure(figsize=(10,10))
plt.scatter(house['sqft_living'],house['price'])
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(house['waterfront'],house['price'])
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(house['zipcode'],house['price'])
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(house['bathrooms'],house['price'])
plt.scatter(house['bedrooms'],house['price'])
plt.show()
```

```python
#histogram
plt.figure(figsize=(10,10))
plt.xlim((0,4000000))
plt.hist(house['price'],bins=30)
plt.show()
```

```python
price_mean = house['price'].mean()
```

```python
price_mean
```

```python
plt.figure(figsize=(10,10))
plt.xlim((0,4000000))
plt.hist(house['price'],bins=30)
plt.axvline(price_mean,color="red")
plt.axvline(house['price'].median(),color="yellow")
plt.legend(labels=['Mean','Median'])
plt.show()
```

```python
count = house['bedrooms'].value_counts().reset_index()
count.rename(columns={'index':'bedroom','bedrooms':'count'},inplace=True)
```
